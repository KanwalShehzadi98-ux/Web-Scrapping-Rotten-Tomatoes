```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://editorial.rottentomatoes.com/guide/essential-movies-to-watch-now"
r = requests.get(url)
r.ok
```

```python
# r.text
```

```python
soup = BeautifulSoup(r.text,"html.parser")
soup.prettify()
```

```python
soup.find("img")
```

```python
soup.find_all("img")
```

```python
divs = soup.find_all("div",{"class" : "col-sm-18 col-full-xs countdown-item-content"})
divs[1].prettify()
```

```python
divs[1].text
```

```python
headings = []
for div in divs:
  headings.append(div.find('h2'))
# print(headings)
for heading in headings:
  print(heading.get_text())
```

```python
movies = []
for movie_name in headings:
  movies.append(movie_name.find('a'))
for movie in movies:
  print(movie.get_text())
```

```python
len(movies)
```

```python
years = []
for year_name in headings:
  years.append(year_name.find("span",{"class":"start-year"}))
for year in years:
  print(year.get_text().strip("()"))
```

```python
len(years)
```

```python
scores = []
for score in headings:
  scores.append(score.find("span",{"class":"tMeterScore"}))
for score in scores:
  print(score.get_text())
```

```python
len(scores)
```

```python
directors = []
for director in divs:
  directors.append(director.find("div",class_="director"))
directors
```

```python
dirs = []
for d in directors:
  dir_link = d.find_all("a")
  dir_name = [dl.text for dl in dir_link]
  result = ','.join(dir_name)
  dirs.append(result)

dirs
```

```python
import pandas as pd
movie_info = pd.DataFrame()
movie_info["Movie_Title"] = movies
movie_info["Year"] = years
movie_info["Score"] = scores
movie_info["Directors"] = dirs
movie_info
```
